```python
#importando librerias necesarias

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/Users/skj/Desktop/PI/PI_ML_OPS/movies_database.csv")
```

```python
# Explora la forma del conjunto de datos
print("El conjunto de datos tiene {} filas y {} columnas".format(df.shape[0], df.shape[1]))

# Examina las primeras y últimas filas del conjunto de datos
print("Las primeras cinco filas del conjunto de datos son:")
print(df.head())

print("Las últimas cinco filas del conjunto de datos son:")
print(df.tail())

# Revisa las estadísticas descriptivas básicas de cada variable
print("Estadísticas descriptivas básicas:")
print(df.describe())
```

```python
# Realiza un análisis visual exploratorio
# Gráfico de barras para variable categórica
sns.countplot(x='type', data=df)
plt.title('Frecuencia de películas y series')
plt.xlabel('Categorías')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Filtrar solo las observaciones con duration_type == 'min'
df_min = df.query('duration_type == "min"')

# Histograma para variable numérica
sns.histplot(x='duration_int', data=df_min, bins=50)
plt.title('Distribución de duración en minutos de películas')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Histograma para variable numérica
sns.histplot(x='score', data=df, bins=50)
plt.title('Distribución del score de películas y series')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Identifica valores atípicos
# Gráfico de caja para variable numérica
sns.boxplot(x='score', data=df)
plt.title('Diagrama de caja para el score')
plt.show()
```

```python
# Filtrar solo las observaciones con duration_type == 'min'
df_min = df.query('duration_type == "min"')

# Identifica valores atípicos
# Gráfico de caja para variable numérica
sns.boxplot(x='duration_int', data=df_min)
plt.title('Diagrama de caja para la duración de películas')
plt.show()
```
